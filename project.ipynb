# 호텔 예약 데이터 컬럼 설명

| 컬럼명                        | 설명 |
|-------------------------------|------|
| **hotel**                     | 호텔의 유형을 나타내며, 예를 들어 'Resort Hotel'이나 'City Hotel' 등으로 구분될 수 있습니다. |
| **is_canceled**               | 예약이 취소되었는지를 나타내는 이진 값 (0: 취소되지 않음, 1: 취소됨). |
| **lead_time**                 | 고객이 예약을 하기까지 걸린 기간을 나타내며, 예약이 접수되기 전의 일수를 의미합니다. |
| **arrival_date_year**         | 고객이 도착하는 연도 (예: 2017). |
| **arrival_date_month**        | 고객이 도착하는 월 (예: 'January', 'February'). |
| **arrival_date_week_number**  | 고객이 도착하는 주차 번호 (1~52 범위). |
| **arrival_date_day_of_month** | 고객이 도착하는 월의 날짜 (1~31 범위). |
| **stays_in_weekend_nights**   | 주말(금, 토) 동안 숙박하는 일수. |
| **stays_in_week_nights**      | 주중(월~목) 동안 숙박하는 일수. |
| **adults**                    | 예약에 포함된 성인 수. |
| **children**                  | 예약에 포함된 어린이 수. |
| **babies**                    | 예약에 포함된 유아 수. |
| **meal**                      | 예약된 식사 유형 (예: 'BB': 조식 포함, 'HB': 조식 및 석식 포함 등). |
| **country**                   | 고객의 거주 국가 코드 (예: 'PRT': 포르투갈, 'GBR': 영국). |
| **market_segment**            | 고객의 예약 경로 (예: 'Direct', 'Corporate', 'Online TA'). |
| **distribution_channel**      | 예약이 들어온 유통 경로 (예: 'Direct', 'TA/TO' (여행사/투어 오퍼레이터)). |
| **is_repeated_guest**         | 고객이 재방문인지 여부 (0: 최초 방문, 1: 재방문). |
| **previous_cancellations**    | 이전 예약 취소 횟수. |
| **previous_bookings_not_canceled** | 이전에 취소되지 않은 예약 횟수. |
| **reserved_room_type**        | 고객이 예약한 방의 유형. |
| **assigned_room_type**        | 체크인 시 할당된 방의 유형 (예약 시 방과 다른 방이 할당될 수 있음). |
| **booking_changes**           | 예약 변경 횟수. |
| **deposit_type**              | 보증금 유형 (예: 'No Deposit': 보증금 없음, 'Non Refund': 환불 불가, 'Refundable': 환불 가능). |
| **agent**                     | 예약을 진행한 에이전트의 ID. |
| **company**                   | 예약한 회사의 ID. |
| **days_in_waiting_list**      | 대기자 명단에서 대기한 일수. |
| **customer_type**             | 고객 유형 (예: 'Transient': 단기 숙박, 'Contract': 계약 고객 등). |
| **adr**                       | 평균 일일 요금 (Average Daily Rate), 총 예약 요금을 숙박일 수로 나눈 값. |
| **required_car_parking_spaces** | 요청된 주차 공간 수. |
| **total_of_special_requests** | 고객의 특수 요청 개수. |
| **reservation_status**        | 예약 상태 (예: 'Check-Out': 체크아웃 완료, 'Canceled': 취소됨). |
| **reservation_status_date**   | 예약 상태가 업데이트된 날짜. |

```python
def clean_channel(row):
    if row['market_segment'] in ['Online TA', 'Offline TA/TO'] and row['distribution_channel'] in ['TA/TO', 'Direct']:
        return 'Travel Agency'
    elif row['market_segment'] == 'Corporate' or row['distribution_channel'] == 'Corporate':
        return 'Corporate'
    elif row['market_segment'] == 'Direct' or row['distribution_channel'] == 'Direct':
        return 'Direct'
    else:
        return row['market_segment']  # fallback

df_hotel['merged_channel'] = df_hotel.apply(clean_channel, axis=1)
```

```python
df_hotel.drop(columns=['market_segment' , 'distribution_channel'], inplace=True)
```

```python
df_hotel = df_hotel[~(df_hotel['merged_channel'] == 'Undefined')]
```

```python
meal_map = {
    'SC': 0,
    'BB': 1,
    'HB': 2,
    'FB': 3,
    'Undefined': 0
}

df_hotel['meal_numeric'] = df_hotel['meal'].map(meal_map)
```

```python
# 1. 관련 컬럼만 추출
corr_cols = ['meal_numeric', 'required_car_parking_spaces', 'total_of_special_requests', 'is_canceled']
corr_df = df_hotel[corr_cols].copy()

# 2. 상관계수 확인
correlations = corr_df.corr()['is_canceled'].drop('is_canceled').abs()
correlations.round(3)
```

```python
# 1. 절댓값 기준 정규화
corr_values = {
    'meal_numeric': 0.004,
    'required_car_parking_spaces': 0.195,
    'total_of_special_requests': 0.236
}

total = sum(corr_values.values())
weights = {k: round(v/total, 3) for k, v in corr_values.items()}
print(weights)
```

```python
from sklearn.preprocessing import StandardScaler

# 1. 서비스 컬럼 지정
service_cols = ['meal_numeric', 'required_car_parking_spaces', 'total_of_special_requests']

# 2. 상관계수 기반 가중치 직접 지정 (이미 계산된 값 사용)
weights = {
    'meal_numeric': 0.009,  
    'required_car_parking_spaces': 0.452,
    'total_of_special_requests': 0.539
}

# 3. 표준화 (StandardScaler)
scaler = StandardScaler()
scaled = scaler.fit_transform(df_hotel[service_cols])

# 4. 표준화된 값을 데이터프레임으로 변환
scaled_df = pd.DataFrame(scaled, columns=['meal_w', 'parking_w', 'requests_w'])

# 5. 가중치 곱하고 total_service 점수 생성
df_hotel['total_service'] = (
    weights['meal_numeric'] * scaled_df['meal_w'] +
    weights['required_car_parking_spaces'] * scaled_df['parking_w'] +
    weights['total_of_special_requests'] * scaled_df['requests_w']
)
df_hotel.drop(columns=['meal_numeric','required_car_parking_spaces','total_of_special_requests'], inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
import pandas as pd

X = df.drop(columns=['is_canceled', 'arrival_date'])
y = df['is_canceled']


# 1. 70:20:10 비율로 나누기 (Stratified 유지)
X_temp, X_test, y_temp, y_test = train_test_split(X, y, test_size=0.1, stratify=y, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=2/9, stratify=y_temp, random_state=42)
# 이유: 2/(7+2) = 2/9 ≈ 0.222 → train:val = 7:2

# 2. 결과 저장 딕셔너리
results = {}

# 3. 평가 함수 정의
def evaluate_model(name, model):
    model.fit(X_train, y_train)
    y_val_pred = model.predict(X_val)
    y_test_pred = model.predict(X_test)

    val_report = classification_report(y_val, y_val_pred, output_dict=True)
    test_report = classification_report(y_test, y_test_pred, output_dict=True)

    results[name] = {
        'val_precision': val_report['weighted avg']['precision'],
        'val_recall': val_report['weighted avg']['recall'],
        'val_f1': val_report['weighted avg']['f1-score'],
        'val_accuracy': val_report['accuracy'],
        'test_precision': test_report['weighted avg']['precision'],
        'test_recall': test_report['weighted avg']['recall'],
        'test_f1': test_report['weighted avg']['f1-score'],
        'test_accuracy': test_report['accuracy']
    }

# 4. 모델 학습 + 평가
evaluate_model('Logistic', LogisticRegression(max_iter=100, class_weight='balanced'))

evaluate_model('Random Forest', RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced'))

pos_ratio = sum(y_train == 0) / sum(y_train == 1)
evaluate_model('XGBoost', XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42, scale_pos_weight=pos_ratio))

evaluate_model('LightGBM', LGBMClassifier(random_state=42, is_unbalance=True))

# 5. 결과 표 출력
df_results = pd.DataFrame(results).T.round(2)
print("\n📋 모델 성능 비교 (Validation & Test):\n")
print(df_results)
```

```python
# 1. 랜덤포레스트 모델 다시 학습
rf_model = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')
rf_model.fit(X_train, y_train)

# 2. 중요도 추출
importances = rf_model.feature_importances_

# 3. 변수 이름과 함께 정렬
feature_importance_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

# 4. 상위 N개 출력
top_n = 20  # 예: 상위 10개
print(f"\n🌟 랜덤포레스트 중요 피처 상위 {top_n}개:")
print(feature_importance_df.head(top_n))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.barplot(
    data=feature_importance_df.head(top_n),
    x='Importance',
    y='Feature',
    palette=sns.color_palette("viridis", n_colors=top_n)  # 여러 색상 지정
)
plt.title("Random Forest - 중요 Feature")
plt.tight_layout()
plt.show()
```
